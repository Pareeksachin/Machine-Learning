```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import KFold,train_test_split,GridSearchCV,RandomizedSearchCV
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
from sklearn.impute import SimpleImputer 
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
```

```python
a = np.array([[2,7,45],
              [4,3,33],
              [np.nan, 1, np.nan],
              [6,np.nan, 56],
              [	np.nan, 7,67],
              [5,4,87]])

imputer = SimpleImputer(strategy='median') 
imputer.fit(a)
print(imputer.statistics_)
imputer.transform(a)
imputer.fit_transform(a)

```

```python

chem = pd.read_csv("..\Cases\Chemical Process Data\ChemicalProcess.csv")
print(chem.isnull().sum())

X = chem.drop('Yield', axis=1)
y = chem['Yield']

X_train,X_test , y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=23)


imputer = SimpleImputer() 
imp_X_train = imputer.fit_transform(X_train)
imp_X_test = imputer.fit_transform(X_test)

poly = PolynomialFeatures(degree=2)

X_train_poly = imputer.fit_transform(imp_X_train)
X_test_poly = imputer.fit_transform(imp_X_test)

lr = LinearRegression()
lr.fit(X_train_poly,y_train)
y_pred = lr.predict(X_test_poly)
print(mean_squared_error(y_test,y_pred))
```

```python
imputer = SimpleImputer()
poly = PolynomialFeatures(degree=2)
lr = LinearRegression()
pipe = Pipeline([('IMPUTER',imputer),('POLY',poly),('LR',lr)])
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
print(mean_squared_error(y_test,y_pred))
```

```python
#### grid Search 
kfold = KFold(n_splits = 5, shuffle=True,random_state = 13)
imputer = SimpleImputer()
poly = PolynomialFeatures(degree=2)
lr = LinearRegression()
pipe = Pipeline([('IMPUTER',imputer),('POLY',poly),('LR',lr)])
print(pipe.get_params())
params = {'IMPUTER__strategy':['mean','median'],'POLY__degree':[1,2,3,4]}
gcv = GridSearchCV(pipe,param_grid=params,verbose = 3,cv = kfold,scoring = 'neg_mean_squared_error')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
################ boston

boston = pd.read_csv(r"C:\Users\Anonymous\Desktop\Advance Statistics\Datasets\boston.csv")
boston
```

```python
X = boston.drop('medv',axis =1 )
y = boston['medv']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=23)
ridge = Ridge(alpha = 1)

ridge.fit(X_train,y_train)
y_cap = ridge.predict(X_test)
print(mean_squared_error(y_test,y_cap))
```

```python
alpha = [0.001,0.01,0.1,0.5,1,1.5,2,5,7,10]
score = []

for i in alpha:
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=23)
    ridge = Ridge(alpha = i)
    ridge.fit(X_train,y_train)
    y_cap = ridge.predict(X_test)
    score.append(mean_squared_error(y_test,y_cap))
print(score)
    
print("Best Score : ",alpha[np.argmin(score)])
#print("Best value ",score[np.argmin])
```

```python
#### kfold 
kfold = KFold(n_splits=5,shuffle=True,random_state=23)
ridge = Ridge()
print(ridge.get_params())
params = {'alpha' : [0.001,0.01,0.1,0.5,1,1.5,2,5,7,10]}
gcv = GridSearchCV(ridge,param_grid=params,verbose = 3,cv = kfold,scoring = 'neg_mean_squared_error')
gcv.fit(X,y)
print(gcv.best_params_)
print(gcv.best_score_)
```

```python
## randomized grid search cv
#### kfold 
kfold = KFold(n_splits=5,shuffle=True,random_state=23)
ridge = Ridge()
print(ridge.get_params())
params = {'alpha' : np.linspace(0.001,15,20)}
Rgcv = RandomizedSearchCV(ridge,param_distributions=params,n_iter=20,verbose = 3,cv = kfold,scoring = 'neg_mean_squared_error')
Rgcv.fit(X,y)
print(Rgcv.best_params_)
print(Rgcv.best_score_)
```
