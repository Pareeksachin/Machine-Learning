```python
import pandas as pd 
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures,LabelEncoder,StandardScaler,MinMaxScaler
from sklearn.model_selection import KFold,train_test_split,GridSearchCV,RandomizedSearchCV, StratifiedKFold
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
from sklearn.impute import SimpleImputer 
from sklearn.metrics import mean_squared_error,log_loss
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
bankrupt = pd.read_csv(r"C:\Users\Anonymous\Desktop\Advance Statistics\Bankruptcy\Bankruptcy.csv")
bankrupt
```

```python
X = bankrupt.drop(['NO','D','YR'],axis=1)
y = bankrupt['D']
```

```python
svm = SVC(kernel='linear',probability=True,random_state=20)
kfold = StratifiedKFold(n_splits=5,shuffle=True,random_state=20)
results = cross_val_score(svm,X,y,cv = kfold,scoring = 'neg_log_loss')
print(results.mean())
```

```python
params = {'C':np.linspace(0.001,5,20)}
gcv = GridSearchCV(svm,param_grid=params,cv = kfold,scoring = 'neg_log_loss',verbose = 3)
gcv.fit(X,y)
```

```python
print(gcv.best_score_)
print(gcv.best_params_)
```
